```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Book1.csv")
df.head()
```

```python
x=df["Country name"]
y=df["Ladder score"]
```

```python
plt.bar(x,y,color="orange")

plt.title("Happiness Index",color="purple")
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

```python


colors=["red","cyan","skyblue","green","white"]
plt.pie(y,labels=x,shadow=True,colors=colors,autopct='%1.1f%%')
plt.title("Happiness Index",color="purple")
plt.show()


```
