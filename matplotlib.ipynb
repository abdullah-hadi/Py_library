```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(21,31)
```

```python
plt.plot(x,x**2,"co-",linewidth=2,markersize=7)
plt.xlabel("X long")
plt.ylabel("Y long")
plt.title("How to complete")
print(x,y)
plt.savefig("NWE.svg")
```

```python
x=np.arange(1,51)
plt.title("try demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,x**2+5*x+4,"r")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,"black")
#plt.axis=("equal")
```

```python
#subplots of sin cos wave
x = np.arange(0,5 * np.pi ,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

plt.savefig('sin_cos_wave.jpg')
plt.show()
```
